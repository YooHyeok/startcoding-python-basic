````python
""" 
[함수]: 특정 작업을 수행하는 코드를 모아놓은 것
```
def 함수명(매개변수1, 매개변수2 , ... ):
  계산로직
  return 결과값 (생략가능)
```
위와 같은 형태로 정의하고 `함수명(매개변수1, 매개변수2, ...)` 형태로 호출한다.
`변수명 = 함수명(매개변수1, 매개변수2, ...)`과 같이 변수에 저장도 가능하다.

사용 이유
1. 반복되는 코드를 줄일 수 있음
2. 재사용하기 편해짐
3. 코드를 수정하기 편해짐
 """

def easy_recipe(a, b):
  print(f"{a}와 {b}을(를) 이용해서 맛있는 요리를 시작합니다.")
  result = a + b + '볶음밥'
  print("요리 완료!")
  return result
````

````python
print("바가 출출하네....")
print("오늘은 야식으로 무엇을 야무지게 먹을까?")
dish = easy_recipe('치즈', '불닭')
dish
````
