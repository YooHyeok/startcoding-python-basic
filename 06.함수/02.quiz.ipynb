```python
""" 
[실습문제 14]
"리그 오브 레전드" 게임에서 세명의 팀원이 얻은 골드량을 매개변수로 받아,  
팀의 평균 골드 획득량을 계산하는 함수 cal_average_gold를 정의하고 새용해보세요.

<소스코드>
cal_average_gold(12000, 15000, 18000)

<실행결과>
15000.0
 """
def cal_average_gold(player1, player2, player3):
  return (player1 + player2 + player3) / 3


cal_average_gold(12000, 15000, 18000)


# 정답
def cal_average_gold(a, b, c):
  avg = (a + b + c) / 3
  return avg


cal_average_gold(12000, 15000, 18000)
```
