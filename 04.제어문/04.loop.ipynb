```python
""" 
[제어문 - for 반복문]

`for 변수 in 순서열` 형태로 사용한다.
순서열: 순서가 있는 숫자열

단순 범위 증가시에는 range() 함수를 사용한다.  
range(a, b) 함수는 정해진 범위 내 순서열을 생성한다.
 """
for i in range(3):
  print("푸들")
  print("포메라니안")
```

```python
""" 
range()의 인수가 1개라면 시작값 0부터 인수 -1까지가 범위로 설정된다.
 """

for i in range(3):
  print(i)
```

```python
for i in range(3):
  print(f"{i + 1}번 과제 시작")
  print("과제 완료")
```

```python
""" 
range()의 인수가 2개라면 시작값 첫번째 인수부터 두번째인수 -1까지가 범위로 설정된다.
 """
for i in range(1, 4): # 1~3
  print(i)
```

```python
""" 
range의 3번째 인자: step - 증가값
 """
for i in range(1, 10, 2): # 1~3 2씩 증가
  print(i)
```
