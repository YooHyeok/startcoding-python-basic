```python
""" 
[실습문제 8]
구구단 출력 프로그램을 만들어 보자.  
프로그램 사용자로부터 출력할 단을 입력받고, 해당 구구단을 출력하는 프로그램이다.  
(for문과 while문 두가지 방식으로 구현해보자)

<입력>
몇 단을 출력할까요?: 5

<출력>
5 * 1 = 5
5 * 2 = 10
5 * 3 = 15
...
5 * 9 = 45
 """
n = int(input("몇 단을 출력할까요?:"))
for i in range(1, 10):
  print(f"{n} * {i} = {n*i}")

i=1
while i < 10:
  print(f"{n} * {i} = {n*i}")
  i += 1


# 정답
x = int(input("몇 단을 출력할까요?:"))
for i in range(1, 10):
  print(f"{x} * {i} = {x*i}")

i=1
while i < 10:
  print(f"{x} * {i} = {x*i}")
  i += 1
```

```python
""" 
[실습문제 9]
로그인 프로그램을 만들어 보자.
프로그램 사용자로부터 비밀번호를 입력받고, 
비밀번호가 맞으면 "로그인 성공!" 을 출력하고,
비밀번호가 틀리면 "로그인 실패!" 를 출력한다.
올바른 비밀번호를 입력할 때까지 계속 반복한다.

<입력>
비밀번호를 입력하세요: 123
비밀번호를 입력하세요: 123
비밀번호를 입력하세요: 123
비밀번호를 입력하세요: 1234

<출력>
로그인 실패!
로그인 실패!
로그인 실패!
로그인 성공!
 """

# 정답
origin_pass = "1234"
while True:
  input_pass = input("비밀번호를 입력하세요:")
  if input_pass == origin_pass:
    print("로그인 성공!")
  else:
    print("로그인 실패!")
```
