```python
""" 
[실습문제 6]
for문을 이용해서 다음과 같은 내용을 출력하는 코드를 작성해 보세요.
<코드 1.>
1번째 강아지
강아지를 위한 장남감 준비 중
장난감 준비 완료

2번째 강아지
강아지를 위한 장남감 준비 중
장난감 준비 완료

3번째 강아지
강아지를 위한 장남감 준비 중
장난감 준비 완료
 """

for i in range(1, 4):
  print(f"{i}번째 강아지")
  print("강아지를 위한 장난감 준비 중")
  print("장난감 준비 완료")
  if i != 3: print()

# 정답
for i in range(1, 4):
  print(f"{i}번째 강아지")
  print("강아지를 위한 장난감 준비 중")
  print("장난감 준비 완료")

```

```python
""" 
[실습문제 7 도전!]
for문을 사용해서 1부터 10까지 합계를 계산해주는 프로그램을 작성해보자.  
힌트) 합계를 저장할 변수(total)를 하나 생성해 준다.
 """
sum = 0
for i in range(1, 11):
  sum+=i
print(sum)

# 정답
sum = 0
for i in range(1, 11):
  sum+=i
print(sum)
```
