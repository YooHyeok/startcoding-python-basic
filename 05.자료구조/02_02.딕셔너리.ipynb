```python
""" 
[딕셔너리]
키(key)와 값(value)를 쌍(pair)으로 가지는 사전형태의 자료형
{} 중괄호 형태로 데이터들을 감싸준다.
`변수 = {'key': value, 'key': value}` 형태로 선언한다.  

딕셔너리 순회 함수
변수.keys(): 딕셔너리의 모든 key를 가져온다.
변수.values(): 딕셔너리의 모든 value를 가져온다.
변수.items(): 딕셔너리의 모든 key와 value를 가져온다.
 """

stock_prices = {
  "삼성전자": 81000,
  "SK하이닉스": 140000,
  "네이버": 35000
}
stock_prices # {'삼성전자': 81000, 'SK하이닉스': 140000, '네이버': 35000}
```

```python
stock_prices['삼성전자'] # 81000
stock_prices['네이버'] # 35000
stock_prices['스타트코딩'] # 35000
```

```python
""" 
딕셔너리 key 목록 순회하기
 """
for key in stock_prices.keys():
  print(key)
```

```python
""" 
딕셔너리 value 목록 순회하기
 """
for value in stock_prices.values():
  print(value)
```

```python
""" 
딕셔너리 item 목록 순회하기
 """
for item in stock_prices.items():
  print(item)
for key, value in stock_prices.items():
  print(key, value)
```
