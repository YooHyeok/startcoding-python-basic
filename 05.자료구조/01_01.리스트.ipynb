```python
""" 
[리스트]
 """
company_list = ["삼성전자", "SK하이닉스", "네이버"]
company_list # 주피터 노트북은 마지막 라인의 변수나  데이터를 자동으로 출력해주는 기능이 있다.
```

```python
""" 
[리스트 - 인덱싱(데이터 접근)]
 """
company_list[0]
company_list[1]
company_list[-1] # 마지막 원소 (0이 첫번째므로)
company_list[3] # [Error] IndexError: list index out of range
```

```python
""" 
[리스트 - 데이터 할당]
 """
company_list[0] = "애플"
company_list[1] = "구글"
company_list[2] = "테슬라"
company_list
```

```python
""" 
[리스트 - 데이터 추가]
 """
company_list.append("마이크로소프트")
company_list # ['애플', '구글', '테슬라', '마이크로소프트']
```

```python
""" 
[리스트 - 데이터 삭제] 
"""
del company_list[0] # 첫번째 인덱스 제거 (애플)
company_list # ['구글', '테슬라', '마이크로소프트']
```

```python
""" 
[리스트 - 길이]
 """
len(company_list)
```

```python

""" 
[리스트 - 슬라이싱(자르기)]
 """
num_list = [1, 2, 3, 4, 5, 6, 7, 8]
num_list[0:2] # [1, 2]: 인덱스 0 부터 2의 앞까지
num_list[:2] # [1, 2]: 첫번째 0은 생략 가능
num_list[1:4] # [2, 3, 4]: 인덱스 1부터 3까지
num_list[3:8] # [4, 5, 6, 7, 8]: 인덱스 3부터 7까지
num_list[3:] # [4, 5, 6, 7, 8]: 마지막 인덱스 생략 가능
num_list[::2] # [1, 3, 5, 7]: 3번째는 step 즉, 2일경우 2단계씩 건너뜀 - 0, 2, 4, 6 인덱스
```
