```python
""" 
[실습문제 12]
여러 회사의 주식 가격이 회사별로 딕셔너리에 리스트로 저장되어 있다.
각 회사별 주식 가격의 평균을 계산하는 프로그램을 작성하세요.

stock_prices = {
  '삼성전자': [81000, 81500, 82000],
  'SK하이닉스': [140000, 141000, 139500],
  '네이버': [350000, 355000, 345000]
}

<실행결과>
# 삼성전자 평균 주식 가격: 81500
# SK하이닉스 평균 주식 가격: 140166
# 네이버 평균 주식 가격: 350000
 """
stock_prices = {
  '삼성전자': [81000, 81500, 82000],
  'SK하이닉스': [140000, 141000, 139500],
  '네이버': [350000, 355000, 345000]
}

for (key, value) in stock_prices.items():
  print(f"# {key} 평균 주식 가격: {int(sum(value) / len(value))}")

# 정답
for (key, value) in stock_prices.items():
  avg = sum(key) / len(value)
  print(f"# {key} 평균 주식 가격: {int(avg)}")
```

```python
""" 
[실습문제 13]
여러 회사의 주식 가격이 회사별로 딕셔너리에 리스트로 저장되어 있다.
각 회사별로 '최고가'와 '최저가'를 출력하는 프로그램을 작성하세요.

stock_prices = {
'삼성전자': {'최고가': 85000, '최저가': 80000, '현재가': 82000},
'SK하이닉스': {'최고가': 145000, '최저가': 139000, '현재가': 140500},
'네이버': {'최고가': 360000, '최저가': 340000, '현재가': 350000},
}

<실행결과>
삼성전자 - 최고가: 85000, 최저가: 80000
SK하이닉스 - 최고가: 145000, 최저가: 139000
네이버 - 최고가: 360000, 최저가: 340000
 """
stock_prices = {
'삼성전자': {'최고가': 85000, '최저가': 80000, '현재가': 82000},
'SK하이닉스': {'최고가': 145000, '최저가': 139000, '현재가': 140500},
'네이버': {'최고가': 360000, '최저가': 340000, '현재가': 350000},
}

max = '최고가'
min = '최저가'
for (key, values) in stock_prices.items():
  print(f"{key} - {max}: {values[max]}, {min}: {values[min]}")

# 정답
for (key, value) in stock_prices.items():
  max_value = value['최고가']
  min_value = value['최저가']
  print(f"{key} - '최고가': {max_value}, '최저가': {min_value}")
```
